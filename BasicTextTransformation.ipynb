```
## NLTK includes a small selection of texts from the Project Gutenberg electronic text archive, which contains some 25,000 free electronic books
import nltk
from nltk.corpus import gutenberg
from pprint import pprint
```

```
nltk.download('gutenberg')
nltk.corpus.gutenberg.fileids()
```

In the next steps, shakespeare-hamlet example will be used. For practive try to replace with any other book in the gutenberg collection.

```
## load the book
hamlet = gutenberg.raw (fileids='shakespeare-hamlet.txt')
## total character in Shakespeare Hamlet
print(len(hamlet))
```

```
nltk.download('punkt')
```

Tokenization and Removing Special Characters

```
tokens = nltk.word_tokenize(hamlet)
type(tokens)
len(tokens)
```

```
tokens[:10] # print first 10 tokens
```

```
## library re help to remove special character like whitespeces and escalation
import re
normalized_tokens = [re.sub(r'\W+', '', token.lower()) for token in tokens]
print(len(normalized_tokens))
```

Stopword Removal

```
from nltk.corpus import stopwords # library
nltk.download('stopwords')
```

```
all_stopwords = set(stopwords.words('english')) # set a language
```

```
# Remove single-character token (mostly punctuation)
normalized_tokens =[normalized_tokens for normalized_tokens in normalized_tokens if len(normalized_tokens) > 1]
```

```
# Remove numbers
normalized_tokens = [normalized_tokens for normalized_tokens in normalized_tokens if not normalized_tokens.isnumeric()]
```

```
# check number of tokens left after the first two pre-processing steps
print(len(normalized_tokens))
```

```
# Remove stopwords
normalized_tokens = [normalized_tokens for normalized_tokens in normalized_tokens if normalized_tokens not in all_stopwords]
print(len(normalized_tokens))
```

Frequency Distribution

```
fdist =  nltk.FreqDist(normalized_tokens)
# output top 10 words
print(fdist.most_common(10))
```

Stemming

```
from nltk.stem import PorterStemmer # used Porter Stemmer, you can use any other stemmer, like snowball
ps = PorterStemmer()
for normalized_tokens in normalized_tokens:
  print(normalized_tokens + ":" + ps.stem(normalized_tokens)) 
```
